```python
import os
import pandas as pd
```

```python
path = os.path.join("Resources", "movie_scores.csv")
df = pd.read_csv(path)
df.head(2)
```

```python
df.columns
```

```python
# only interested in IMDb data, so create a new table that takes the Film and all the columns relating to IMDB
df[["FILM", "IMDB", "IMDB_norm", "IMDB_norm_round", "IMDB_user_vote_count"]]
```

```python
# Filter out only the good movies—i.e., any film with an IMDb score greater than or equal to 7 and remove the norm ratings
df[df["IMDB"] >= 7]
```

```python
# Find less popular movies that you may not have heard about - i.e., anything with under 20K votes
df.loc[df["IMDB"] >= 7, ["FILM", "IMDB", "IMDB_norm", "IMDB_norm_round", "IMDB_user_vote_count"]]
df[df["IMDB_user_vote_count"] < 20000]
# df = df.loc[df["IMDB"] >= 7 & (df[IMDB_user_vote_count"] < 20000), ["FILM", "IMDB", "IMDB_norm", "IMDB_norm_round", "IMDB_user_vote_count"]]
df.sort_values("IMDB", ascending=False).head(3)
```

```python
# Finally, export this file to a spreadsheet, excluding the index, so we can keep track of our future watchlist
df.to_csv("GoodFilms.csv", index=False)
df.to_excel("GoodFilms.xlsx", index=False)
# No module named 'openpyxl'
# Go to Pandas_Practice, run "pipenv install openpyxl", then works
```
